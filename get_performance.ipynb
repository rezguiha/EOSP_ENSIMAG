```python
import pickle 
import os
import matplotlib.pyplot as plt
import numpy as np
import glob

collection='AP88-89'
model_name='DIR'
total_loss_folds=[]
ndcg5_folds=[]
ranking_loss_folds=[]
sparsity_folds=[]
proportion_folds=[]
min_number_of_epochs=10000
sparsity_list=[]
for fold in range(5):
    files=glob.glob('TREC/'+collection+'/plots/fold'+str(fold)+'/'+collection+'_'+model_name+'_1e-5_0.0_GPU_weights_2_layers_he_norm_epoch*')
    file_path=files[-1]
    plot_values = pickle.load(open(file_path,'rb'))
    total_loss=[]
    ndcg5=[]
    for i in range(len(plot_values[model_name][2])):
        total_loss.append(plot_values[model_name][2][i])
        ndcg5.append(plot_values[model_name][1][i]['ndcg_cut_20'])
    total_loss_folds.append(total_loss)
    ndcg5_folds.append(ndcg5)
    ranking_loss_folds.append(plot_values[model_name][2])
    sparsity_folds.append(plot_values[model_name][3])
    proportion_folds.append(plot_values[model_name][0])
    if len(plot_values[model_name][2])<min_number_of_epochs:
        min_number_of_epochs=len(plot_values[model_name][2])
    files=glob.glob('TREC/'+collection+'/weights/fold'+str(fold)+'/'+collection+'_'+model_name+'_1e-5_0.0_GPU_weights_2_layers_he_norm/epoch*')
    sparsity_number=[]
    for file_path in files:
        with open(file_path, 'rb') as f1:
            weights = pickle.load(f1)
        number_of_zeros=0
        for weight in weights:
            if weight==0:
                number_of_zeros+=1
        sparsity_number.append(number_of_zeros)
    sparsity_list.append(sparsity_number)
total_loss_folds=[total_loss_folds[i][:min_number_of_epochs] for i in range(5)]
ndcg5_folds=[ndcg5_folds[i][:min_number_of_epochs] for i in range(5)]
ranking_loss_folds=[ranking_loss_folds[i][:min_number_of_epochs] for i in range(5)]
sparsity_folds=[sparsity_folds[i][:min_number_of_epochs] for i in range(5)]
proportion_folds=[proportion_folds[i][:min_number_of_epochs] for i in range(5)]
sparsity_list=[sparsity_list[i][:min_number_of_epochs] for i in range(5)]


proportion_folds=np.array(proportion_folds)
total_loss_folds=np.array(total_loss_folds)
ndcg5_folds=np.array(ndcg5_folds)
ranking_loss_folds=np.array(ranking_loss_folds)
sparsity_folds=np.array(sparsity_folds)

sparsity_number_array=np.array(sparsity_list)

proportion=np.sum(proportion_folds,axis=0)/5
total_loss=np.sum(total_loss_folds, axis=0)/5
ndcg5=np.sum(ndcg5_folds,axis=0)/5
ranking_loss=(np.sum(ranking_loss_folds,axis=0)*(1-0.00001))/5
sparsity=(np.sum(sparsity_folds,axis=0)*0.00001)/5

sparsity_number=np.sum(sparsity_number_array,axis=0)/5

print("ndcg20 max =", max(ndcg5))
for i in range(len(ndcg5)):
    if ndcg5[i]==max(ndcg5):
        print("position max =",i)
        print("proportion at max =",proportion[i])
for i in range(len(ndcg5)):
    if ndcg5[i]<0.339:
        print("position no training =",i)
        print("proportion no training =",proportion[i])            

proportion=np.insert(proportion,0,1.0)
ndcg5=np.insert(ndcg5,0,0.339)

epochs=range(1,min_number_of_epochs+1)
fig, ax = plt.subplots()
ax.set_xlabel('epochs')
ax.set_ylabel('total_loss')
lns1=ax.plot(epochs, total_loss, 'r', label='Total loss')
lns2=ax.plot(epochs, ranking_loss, 'k--', label='Ranking loss')
lns3=ax.plot(epochs, sparsity, 'k:', label='Sparsity term')
ax2 = ax.twinx()
ax2.set_ylabel("ndcg20")
epochs=range(0,min_number_of_epochs+1)
lns4=ax2.plot(epochs, ndcg5, 'b', label='Evaluation ndcg20')
lns5=ax2.plot(epochs,proportion,'g', label='Proportion of elements')
lns = lns1+lns2+lns3+lns4+lns5
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc='best')
# ax.legend(lns, labs,bbox_to_anchor=(0.5, 1), loc=2, borderaxespad=0)
fig.tight_layout()
ax.set_title("Evolution of the loss and the ndcg20 for L1 weight=1e-5 for the model "+model_name+" for the collection "+collection)

# Printing weights of hidden layer
weights=plot_values[model_name][6]

min_weights=[min(weights[i][0]) for i in range(len(weights))]
max_weights=[max(weights[i][0]) for i in range(len(weights))]
average_weights=[sum(weights[i][0])/len(weights[i][0]) for i in range(len(weights))]

fig3, ax3 = plt.subplots()
ax3.set_xlabel('batch')
ax3.set_ylabel('Weights')
lns1=ax3.plot(range(len(min_weights)), min_weights, 'r', label='Minimum weights')
lns2=ax3.plot(range(len(max_weights)), max_weights, 'g', label='Maximum weights')
lns3=ax3.plot(range(len(average_weights)), average_weights, 'k', label='Average weights')

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax3.legend(lns, labs, loc=0)
fig3.tight_layout()
ax3.set_title("Evolution of weights for L1 weight=1e-5 for the model "+model_name+" for the collection "+collection)
#printing bias evolution
weights=plot_values[model_name][6]

bias=[weights[i][1] for i in range(len(weights))]

fig4, ax4 = plt.subplots()
ax4.set_xlabel('batch')
ax4.set_ylabel('Weights')
lns1=ax4.plot(range(len(bias)), bias, 'r', label='Bias')


lns = lns1
labs = [l.get_label() for l in lns]
ax4.legend(lns, labs, loc=0)
fig4.tight_layout()
ax4.set_title("Evolution of the bias for L1 weight=1e-5 for the model "+model_name+" for the collection "+collection)

#printing gradients evolution
gradients=plot_values[model_name][7]

min_gradients=[min(gradients[i][0].numpy()) for i in range(len(gradients))]
max_gradients=[max(gradients[i][0].numpy()) for i in range(len(gradients))]
average_gradients=[sum(gradients[i][0].numpy())/len(gradients[i][0].numpy()) for i in range(len(gradients))]

fig5, ax5 = plt.subplots()
ax5.set_xlabel('batch')
ax5.set_ylabel('Gradients')
lns1=ax5.plot(range(len(min_gradients)), min_gradients, 'r', label='Minimum gradients')
lns2=ax5.plot(range(len(max_gradients)), max_gradients, 'g', label='Maximum gradients')
lns3=ax5.plot(range(len(average_gradients)), average_gradients, 'k', label='Average gradients')

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax5.legend(lns, labs, loc=0)
fig5.tight_layout()
ax5.set_title("Evolution of gradients for L1 weight=1e-5 for the model "+model_name+" for the collection "+collection)

#printing sparsity
fig6, ax6 = plt.subplots()
ax6.set_xlabel('epochs')
ax6.set_ylabel('number of null elements')
lns1=ax6.plot(range(len(sparsity_number)), sparsity_number, 'r', label='number of null TDV weights')

lns = lns1
labs = [l.get_label() for l in lns]
ax6.legend(lns, labs, loc=0)
fig6.tight_layout()
ax6.set_title("Evolution of for the number of null TDVs in the vocabulary for L1 weight=1e-5 for the model "+model_name+"for the collection "+collection)
```
